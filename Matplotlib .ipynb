```python
import numpy as np
import matplotlib as mp
import matplotlib.pyplot as plt
%matplotlib inline
```

Generating Sin and Cos graph

```python
x = np.linspace(0,4*np.pi,100)
sin_y =np.sin(x)
cos_y =np.cos(x)
plt.plot(x,sin_y,label ='Sine')
plt.plot(x,cos_y,label ='Cosine')
plt.legend(loc ='upper right')
plt.title('Sine Wave')
plt.show()
```

```python
plt.plot(x,sin_y,'*')
plt.plot(x,cos_y,'--')
plt.title('Sine Wave')
plt.show()
```

Increase size of plot
plt.figure(figsize=(length,height))

```python
plt.figure(figsize =(18,5))
plt.style.use('grayscale')
plt.plot(x,sin_y,label ='Sine')
plt.plot(x,cos_y,label ='Cosine')
plt.title('Sine & Cosine Wave')
plt.legend()
plt.show()
```

Saving the Plots in our Local Machine

```python
plt.figure(figsize = (18,5))
fig = plt.figure()
plt.plot(x, sin_y,label = 'Sine')
plt.plot(x, cos_y,label = 'Cosine')
plt.title('Sine & Cosine Wave')
fig.savefig('Sine & Cosine.png')
```

Types of Plots

line plot

```python
x = np.linspace(0,4*np.pi, 100)
y = np.sin(x)
plt.style.use('Solarize_Light2')
plt.plot(x, y)
plt.title('Sine & Cosine Wave')
plt.show()
```

Multiple line Plots

```python
x = np.linspace(0,4*np.pi, 100)
y_sin = np.sin(x)
y_cos = np.cos(x)
plt.figure(figsize = (15,5))
plt.style.use('Solarize_Light2')
plt.plot(x, y_sin,  '*' ,color = 'red' , linewidth = 1, label = 'sin')
plt.plot(x, y_cos, '--' , color = 'blue', linewidth = 1, label = 'cos')
plt.title('Sine & Cosine Wave')
plt.legend()
plt.show()
```

We can also use linestyle command. It has 4 main styles solid, dashed, dashdot and dotted

```python
x = np.linspace(0, 20, 21, dtype = 'int')
plt.figure(figsize = (15,8))
plt.plot(x, x + 0, linestyle = 'solid'  , label = 'Solid')
plt.plot(x, x + 1, linestyle = 'dashed' , label = 'Dashed') 
plt.plot(x, x + 2, linestyle = 'dashdot', label = 'DashDot')
plt.plot(x, x + 3, linestyle = 'dotted' , label = 'Dotted')
plt.legend(loc = 'lower right')
plt.show()
```

Pie Charts

```python
cars = ['AUDI','BMW','FORD','TESLA','JAGUAR','MERCEDES']
data = [ 450  , 385 , 35   ,   10  ,  200   , 500      ]
fig = plt.figure(figsize = (7,7))
plt.pie(data, labels = cars)
plt.show()
```

```python
cars = ['AUDI','BMW','FORD','TESLA','JAGUAR','MERCEDES']
data = [ 450  , 385 , 35   ,   10  ,  200   , 500      ]
explode = [0, 0, 0 , 0, .1, 0] 
fig = plt.figure(figsize = (7,7))
plt.pie(data, labels = cars, explode = explode)
plt.show()
```

Adding Shadows

```python
cars = ['AUDI','BMW','FORD','TESLA','JAGUAR','MERCEDES']
data = [ 450  , 385 , 35   ,   10  ,  200   , 500      ]
explode = [0, 0, 0 , 0, .1, 0] 
fig = plt.figure(figsize = (7,7))
plt.pie(data, labels = cars, explode = explode, shadow = True)
plt.show()
```

Adding Percentage

```python
cars = ['AUDI','BMW','FORD','TESLA','JAGUAR','MERCEDES']
data = [ 450  , 385 , 35   ,   50  ,  200   , 500      ]
explode = [0, 0, 0 , 0, .1, 0] 
fig = plt.figure(figsize = (7,7))
plt.pie(data, labels = cars, explode = explode, shadow = True, autopct = '%2.2f%%')
plt.show()
```

Adding Wedget

```python
fig = plt.figure(figsize = (7,7))
plt.pie(data, labels = cars, explode = explode, shadow = True, autopct = '%2.2f%%',
        wedgeprops = {'edgecolor' : 'black', 'linewidth' : 2, 'antialiased' : True})
plt.show()
```

Multiple Subplots

```python
x = np.linspace(0,10,50)
fig, ax = plt.subplots(2,2)
y = np.sin(x)
ax[0,0].plot(x,y)
ax[0,0].set_title('Sin Wave')
y = np.cos(x)
ax[0,1].plot(x,y)
ax[0,1].set_title('Cos Wave')
y = np.tan(x)
ax[1,0].plot(x,y)
ax[1,0].set_title('Tan Wave')
y = np.random.randn(1000)
ax[1,1].hist(y,bins = 40)
ax[1,1].set_title('Random Distribution')
plt.tight_layout()
plt.show()
```

With loops

```python
plt.subplots_adjust(hspace = .4, wspace = .4)
for i in range(1,7):
    plt.subplot(2,3,i)
    plt.text(0.5 , 0.5, str((2,3,i)), fontsize = 18, ha = 'center')
```

```python
x = np.linspace(0,10,50)
y = np.sin(x)
fig , (ax1,ax2,ax3) = plt.subplots(1,3, figsize = (18,6))
ax1.plot(x,y)
ax2.plot(x,y**2)
ax3.plot(x,y**3)
ax1.set_title('Sin Wave')
ax2.set_title('Sin Wave - Square')
ax3.set_title('Sin Wave - Cube')
plt.tight_layout()
plt.show()
```
